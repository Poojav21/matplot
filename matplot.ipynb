```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0, 10, 1)
y = 2*x + 5
y1 = 5*x + 6
fig = plt.figure(figsize=(10, 5))
# plotting the datapoints
plt.plot(x, y,linewidth = 2.0,linestyle = ":", color = "r", alpha = 1, marker = "o" )
plt.plot(x, y1,linewidth = 2.0,linestyle = ":", color = "b", alpha = 1, marker = "o" )
plt.title("Line Plot Demo")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["lines1", "line2"], loc = "best")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0, 10, 1)
y1 = 2*x + 5
y2 = 3*x + 5
plt.subplot(1,2,1)
# creating the points
plt.plot(x, y1)
plt.xlabel("X axis")
plt.ylabel("y1 axis")
plt.title("Graph1")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.xlabel("X axis")
plt.ylabel("y2 axis")
plt.title("Graph2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# creating the dataset
data = {"apple":20, "mangoes":15, "Lemon":30, "oranges":10}
names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(names, values, color = "r", alpha = 1)
plt.xlabel("names")
plt.ylabel("values")
plt.title("Price vs Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# creating the dataset
a = [10, 20, 30, 40, 50, 60, 70, 80]
b = [5, 3, 2, 5, 6, 1, 4, 2]
x = [1, 2, 3, 4, 3, 2, 1, 2]
# creating the scatter plot s= 200 (the size of dots), plt.savefig-to save the fig output
plt.scatter(a, b, color = "g", s = 200, edgecolors = "y", marker = "o", alpha = 0.5)
plt.scatter(a, x, color = "r", s = 200, edgecolors = "b", marker = "4", alpha = 1)
plt.title("Scatter plot demo")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.savefig("Scatter_plot.png")
plt.show()
```

```python
#  creating the historgram
# histogram denotes frequency of datas in a range like in this graph b/w 0 t0 20 there are 5 five values in.
import matplotlib.pyplot as plt
%matplotlib inline 
numbers = [10, 12, 16, 19, 11, 20, 26, 28, 30, 38, 35, 34, 45, 60, 64, 62, 70, 78, 75, 79, 85, 94, 95]
# creating the histogram
plt.hist(numbers, bins = [0, 20, 40, 60, 80, 100], color = "r")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
numbers = [10, 12, 16, 19, 11, 20, 26, 28, 30, 38, 35, 34, 45, 60, 64, 62, 70, 78, 75, 79, 85, 94, 95]
# creating the histogram
plt.hist(numbers, bins = [0, 20, 40, 60, 80, 100], edgecolor = "black", color = "g")
plt.title("Histogram demo")
plt.xlabel("Range of values")
plt.ylabel("Frequency of values")
plt.grid(True)
plt.savefig("Histogarm plot.png")
plt.show()
```

```python
# Box plot
import matplotlib.pyplot as plt
%matplotlib inline
# Data preparation
total = [20, 4, 1, 30, 20, 10, 20, 70, 30 ,10]
order = [10, 3, 1, 15, 17, 2, 30, 44, 2, 1]
discount = [30, 10, 20, 5, 10, 20, 50, 60, 20, 45]
data = list([total, order, discount])
# plotting the data
plt.boxplot(data, showmeans=True)
plt.title("Box plot demo")
plt.grid(True)
plt.show()
```

```python
# violin plot
import matplotlib.pyplot as plt
%matplotlib inline
# Data preparation
total = [20, 4, 1, 30, 20, 10, 20, 70, 30 ,10]
order = [10, 3, 1, 15, 17, 2, 30, 44, 2, 1]
discount = [30, 10, 20, 5, 10, 20, 50, 60, 20, 45]
data = list([total, order, discount])
# plotting the data
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title("Violin plot demo")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
label = ["Dog", "Cat", "wolf", "Lion"]
sizes = [50, 45, 60, 80]
plt.pie(sizes, labels = label, autopct="%1.1f%%", shadow = True, startangle=90, explode=(0.1, 0.1, 0, 0))
plt.title("Pie chart demo")
plt.show()
```

```python
# Box plot
import matplotlib.pyplot as plt
%matplotlib inline
# Data preparation
group_names = ["GroupA", "GroupB", "GroupC"]
group_size = [20, 30, 50]
size_centre= [5]
a, b, c = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
pie1 = plt.pie(group_size, labels = group_names, radius = 1.5, colors = [a(0.5), b(1.5), c(1.0)])
pie2 = plt.pie(size_centre, radius = 1.0, colors = "w")
plt.show()
```
